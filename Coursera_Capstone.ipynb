<h3>Scrapping the table from wiki and converting into xlsx file</h3>

```python
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
table = pd.read_html(url, keep_default_na=False) 
df = table[0]
df.to_excel("output.xlsx") 
```

**There were no strict intructions to the methods to be used for data cleaning, so due to speed and simplicity the data was changed in excel and returned as a new dataframe**

```python
data = pd.read_excel('newdata.xlsx')
```

<h3>Importing geospatial data</h3>

```python
geocor = pd.read_csv('Geospatial_Coordinates.csv')
geocor.head()
```

<h3>Merging Tables (adding geo data)</h3>

```python
mergeddata = pd.merge(data, geocor, on='Postal Code')
```

```python
mergeddata.head()
```

```python
mergeddata.shape
```

<h3>Map of Toronto Neighborhoods</h3>

```python
import folium
map_toronto = folium.Map(location=[43.651070, -79.347015], zoom_start=10)
for lat, lng, borough, neighborhood in zip(mergeddata['Latitude'], mergeddata['Longitude'], mergeddata['Borough'], mergeddata['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)
map_toronto
```

**Let's cluster only the neighborhoods that contain "Toronto". So let's slice the original dataframe and create a new dataframe**

```python
data2 = mergeddata[mergeddata['Borough'].str.contains("Toronto")].reset_index(drop=True)
data2.head()
```
